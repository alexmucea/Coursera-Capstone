```python
import pandas as pd
```

```python
# importing csv from previous notebook
data = pd.read_csv('toronto.csv')
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.head()
```

```python
# importing csv containing geospatial coordinates and renaming one column for merging
coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
coordinates.rename(columns={'Postal Code': 'Postcode'},inplace=True)
coordinates.head()
```

```python
# merging to create final dataframe
tdf = pd.merge(data,coordinates,on='Postcode')
tdf.head()
```

```python
# exporting to csv for further steps
tdf.to_csv('toronto_geo.csv')
```
